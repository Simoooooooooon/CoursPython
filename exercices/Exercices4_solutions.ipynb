<u>Exercice4:</u> *Temperatures partie4.*  
- Ajouter une anotation montrant le point de temperature maximale 
- Entourer ce point d'un cercle (ou ellipse) rouge rempli de rose
- tracer une ligne horizontale pour montrer la temperature moyenne la plus haute et une autre pour la temperature moyenne la plus basse.

```python
mois  = ['janv','fev','mar','avr','mai','juin','juil','aout','sept','oct','nov','dec']
Tmoy = [3.9, 9, 10.9, 12.3, 19.7, 23.1 , 25.7, 26.1, 20.4, 19.2, 11.7 , 8.8]

Tmini = [-5, -1.3, -0.2, -1.4, 10, 12.6, 12.2, 15.8, 8.2, 9.0, 1.5, -5.9]
Tmaxi = [16.7, 18.9, 21.5, 23.1, 33.4, 38.9, 39.4, 39.8, 34.4, 29.6, 21.7, 19.7]


fig,ax = plt.subplots()
ax.plot(mois,Tmoy,color = 'green')

ax.plot(mois,Tmini,color = 'blue')
ax.plot(mois,Tmaxi,color = 'red')

ax.set_facecolor( '#FBFBDD')



# Ajouter des noms aux axes et un titre au graphique
ax.set_xlabel("Mois")
ax.set_ylabel("Température moyenne")
ax.set_title("Températures mensuelles moyennes")

ax.grid(True)

# Remplissage de l'aire entre les deux courbes
ax.fill_between(mois, Tmini, Tmaxi, color='orange', alpha=0.3)


# Ajouter une anotation montrant le point de temperature maximale 
index_max = Tmaxi.index(max(Tmaxi))# le point maxi a pour coordonnées
xmax = index_max 
ymax = Tmaxi[index_max]

ax.annotate("Température\n Mensuelle\n Maximale", xy=(xmax, ymax), xytext=(xmax+1, ymax-3),
            arrowprops=dict(arrowstyle='->'), fontsize=10, color='red')


# Entourer ce point d'un cercle (ou ellipse) rouge rempli de rose
radius = 1
circle = plt.Circle((xmax,ymax), radius, edgecolor='red', facecolor='pink')
ax.add_patch(circle)


# tracer une ligne horizontale pour montrer la temperature moyenne la plus haute et une autre pour la temperature moyenne la plus basse.
ymoy_basse = min(Tmoy)
ymoy_haute = max(Tmoy)

ax.axhline(y= ymoy_basse, color='#27E2F5', linestyle='--')
ax.axhline(y= ymoy_haute, color='#F52773', linestyle='--')
```

<u>Exercice 6:</u> *Progressions des notes.*  
- Tracer un graphique montant l'évolution sur le semestre des notes des éléves suivants: 

| Prénom  | Notes         |Date                                       |
|---------|---------------|-------------------------------------------|
| Alice   | 15, 18, 14, 15|14 janvier , 16 avril , 10 mai , 16 juin   |
| Bob     | 12, 16, 17, 13|14 janvier , 16 avril , 10 mai , 16 juin   |
| Charlie | 20, 19, 8 ,15 |14 janvier , 16 avril , 10 mai ,  16 juin  |

- Ajouter un titre et le nom des axes
- Ajouter une legende pour chaque éléve
- Tracer une ligne horizontale en pointillé représentant la moyenne à 10
- Faire un affichage sur un graphiques avec 3 sous graphiques


```python
import matplotlib.pyplot as plt

# Dates
dates = ['14 janvier' , '16 avril' , '10 mai' , '16 juin' ]

# Notes
Alice = [15, 18, 14, 15]
Bob = [12, 16, 17, 13]
Charlie = [20, 19, 8, 15]

# Tracer un graphique montant l'évolution sur le semestre des notes des éléves
fig,ax = plt.subplots()

ax.plot(dates, Alice, label = 'Alice')
ax.plot(dates, Bob, label = 'Bob')
ax.plot(dates, Charlie, label = 'Charlie')

# Ajouter une legende pour chaque éléve (affiche les labels)
ax.legend()

# Ajouter un titre et le nom des axes
ax.set_title('Notes en fonction des dates')
ax.set_xlabel('Dates')
ax.set_ylabel('Notes')

# Tracer une ligne horizontale en pointillé représentant la moyenne
ax.axhline(y=10, color='r', linestyle='--')

fig,ax = plt.subplots(1,3,figsize=(12, 4))
ax[0].plot(dates, Alice, label = 'Alice')
ax[1].plot(dates, Bob, label = 'Bob')
ax[2].plot(dates, Charlie, label = 'Charlie')
ax[0].legend()
ax[1].legend()
ax[2].legend()
ax[0].axhline(y=10, color='r', linestyle='--')
ax[0].axhline(y=10, color='r', linestyle='--')
ax[1].axhline(y=10, color='r', linestyle='--')
ax[2].axhline(y=10, color='r', linestyle='--')


plt.tight_layout()
```

<u>Exercice7:</u> *Puissance*  
- Ecrire une fonction 'liste_croissante' qui crée une liste d'éléments croissants prenant en argument debut, fin, pas
- Ecrire une fonction 'puissance' qui éléve à un puissance passée en argument les éléments d'une liste, également passée en argument.
- Ecrire une fonction 'graphique' qui ajoute a une figure et axes existant un tracé. C'est a dire que la fonction prendra en argument, non seulement las valeurs à tracer, mais aussi la figure et l'axe, ainsi que la puissance à laquelle est élevée la liste.
- Utiliser, dans une boucle, les 3 fonctions définies précedemment pour:
    - Créer une liste de nombres de 1 à 100 avec un pas de 2
    - Mettre successivement cette liste à l'exposant 1 à 10
    - Tracer sur un même graphe une représentation des valeurs de la liste aux exposants de 1 à 10
- Ce graphe est il facilement lisible? Comment l'améliorer?
    

```python
import matplotlib.pyplot as plt

# Ecrire une fonction 'liste_croissante' qui crée une liste d'éléments croissants prenant en argument debut, fin, pas
def liste_croissante(debut, fin, pas):
    liste = []
    for i in range (debut, fin, pas):
        liste.append(i)
    return liste
 
# Ecrire une fonction 'puissance' qui élève à une puissance les éléments d'une liste
def puissance(liste,exposant):
    puissance_liste = []
    for element in liste:     
        puissance_exposant_de_nombre = element**exposant
        puissance_liste.append(puissance_exposant_de_nombre)
    return puissance_liste

# Ecrire une fonction 'graphique' qui ajoute a une figure et axes existant un tracé. C'est a dire que la fonction prendra 
# en argument, non seulement las valeurs à tracer, mais aussi la figure et l'axe. , ainsi que la puissance à laquelle 
# est élevée la liste.
def graphique(liste,puissance_liste,figure,axe,puissance):
    ax.plot(liste,puissance_liste,label = str(puissance))
    ax.legend()
    ax.set_xlabel('Valeur')
    ax.set_ylabel('Puissance de Valeur')
  
    

    
    
```

```python
# - Utiliser, dans une boucle, les 3 fonctions définies précedemment pour:
#     - Créer une liste de nombres de 1 à 100 avec un pas de 2
#     - Mettre successivement cette liste à l'exposant 1 à 10
#     - Tracer sur un même graphe une représentation des valeurs de la liste aux exposants de 1 à 10

fig,ax = plt.subplots()
for n in range (0,10):
    a = liste_croissante(1,100,2)
    a_puissance_n= puissance(a,n)
    graphique(a,a_puissance_n,fig,ax,n)
```

```python
# - Ce graphe est il facilement lisible? Comment l'améliorer?
# --> Mettre l'axe des y en echelle log

fig,ax = plt.subplots()
for n in range (0,10,1):
    a = liste_croissante(1,100,2)
    a_puissance_n= puissance(a,n)
    graphique(a,a_puissance_n,fig,ax,n)
ax.set_yscale('log')
```
