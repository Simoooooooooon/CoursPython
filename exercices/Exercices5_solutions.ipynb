<u>Exercice 1:</u> *Déclaration et valeurs.*  
- Déclarer un tableau numpy de 3 colonnes et 4 lignes, rempli de valeurs aléatoires entre 0 et 50
- Convertir les éléments de ce tableau en nombres flottants codés sur 16 bits
- Afficher le nombre de dimensions du tableau et les valeurs de ces dimensions.
- Multiplier les valeurs de éléments du tableau par 2
- Donner la moyenne des valeurs des éléments du tableau et leur écart type
- Changer les dimensions du tableau pour avoir 2 lignes et 6 colonnes 


```python
import numpy as np

# Déclaration du tableau de 3 colonnes et 4 lignes avec des valeurs aléatoires
tableau = np.random.randint(0, 50, size=(4, 3))
print('tableau déclaré',tableau)

# Conversion des éléments de ce tableau en nombres flottants codés sur 16 bits
tableau =  tableau.astype(np.float16)
print('tableau float 16 bits',tableau)


# Affichage du  nombre de dimensions du tableau 
print("Nombre de dimensions du tableau :", tableau.ndim)
# et les valeurs de ces dimensions
print("Dimensions du tableau :", tableau.shape)


# Multiplication de toutes les valeurs du tableau par 2
tableau  = tableau*2
print(tableau)

# Calcul de la moyenne des valeurs des éléments du tableau
moyenne = np.mean(tableau)
print("Moyenne des valeurs :", moyenne)
# et de l'écart type
ecart_type = np.std(tableau)
print("Ecart type des valeurs :", ecart_type)


# Modification des dimensions du tableau pour obtenir 2 lignes et 6 colonnes
tableau = tableau.reshape(2, 6)

# Affichage du tableau après la modification des dimensions
print("Tableau après modification des dimensions :\n", tableau)


```

<u>Exercice 2:</u> *Indexation et conditions*  
- Déclarer un tableau numpy de 4 colonnes et 4 lignes, rempli de valeurs aléatoires entre 0 et 100
- afficher le tableau avec 1 décimale
- Extraire la première ligne du tableau et assignez-les à une nouvelle variable appelée "sous_tableau". Afficher le contenu du sous-tableau extrait.
- Même question pour la dernière colonne du tableau
- Calculer la somme des éléments de chaque colonne
- Donner le numéro de ligne et de colonne de chaque élement dont la valeur est supérieure à 50

```python

import numpy as np

# Déclarer un tableau numpy de 4 colonnes et 4 lignes, rempli de valeurs aléatoires entre 0 et 100
tableau = np.random.uniform(0, 100, size=(4, 4))
print('Tableau origine:\n',tableau)

# afficher le tableau avec 1 décimale
tableau = np.round(tableau,1)
print('\nTableau avec 1 décimale:\n',tableau)



# Extraction de la première ligne du tableau modifié
sous_tableau = tableau[0]
print("\nPremiere ligne :\n", sous_tableau)

# Extraction de dernière colonne du tableau modifié
sous_tableau = tableau[:,-1]
print("\nDerniere colonne :\n", sous_tableau)

# Calculer la somme des éléments de chaque colonne
somme_col = np.sum(tableau,axis = 0)
print('\nSomme des elements de chaque colonne',somme_col)

# Donner le numéro de ligne et de colonne de chaque élement dont la valeur est supérieure à 50

indices_elements_sup50 = np.where(tableau>50)
print('\nIndices des elements superieurs à 50',indices_elements_sup50)

valeurs_elements_sup50 = tableau[indices_elements_sup50]
print('\nValeur des elements superieurs à 50',valeurs_elements_sup50)
```

Remarque: Pour les indices donnés par np.where,on obtient deux tableaux: le premier pour les numeros de ligne, le second pour les numeros de colonnes. 

#### <u>Exercice 3:</u> *interpolation et dérivée* 
- creer un tableau numpy appelé X de dimension 1, contant des valeurs croissantes de 1 à 10 par pas de 1
- creer un tableau numpy appelé Y de dimension 1, contenant le même nombre d'éléments que X avec des valeurs aléatoires entre 0 et 10.
- Représenter Y en fonction de X sur un graphique. 
- Définir un tableau Y2 contenant les valeurs de Y pour X2 = 0.5, X = 1.5, ....X = 9.5
- Représenter les points de coordonnées X2,Y2 sur le  graphique. 
- Fusionner d'abord les tableaux X et X2 puis Y et Y2, c'est à dire créer un tableau dont les indices pairs seront les valeurs de X et les indices impairs ceux de X2. Idem pour Y et Y2.
- Vérifier en faisant le graphique des tableaux fusionnés
- Tracer la dérivée des Y fusionnés par rapport aux X fusionnés
- Donner l'ordonnée du point 0.25 par une interpolation

```python
import numpy as np
import matplotlib.pyplot as plt

# creer un tableau numpy de dimension 1, contant des valeurs croissantes de 0 à 10 par pas de 1
X = np.arange(11)
print('X:\n',X)
#####################################################
# creer un tableau numpy appelé Y de dimension 1, 
# contant le même nombre d'éléments que X avec des valeurs aléatoires entre 0 et 10.
taille = len(X)
Y = np.random.randint(0, 10, size=(taille))
print('Y:\n',Y)
#####################################################
# Représenter Y en fonction de X sur un graphique
fig,ax = plt.subplots()
ax.plot(X,Y,marker = 'o',markeredgecolor='blue')
ax.set_xlabel('X')
ax.set_ylabel('Y')
#####################################################
# Definir un tableau contenant les valeurs de Y pour X = 0.5, X = 1.5, ....X = 9.5
X2 = np.arange(0.5,10.5,1)
print('X2:\n',X2)
Y2 = np.interp(X2, X, Y)
print('Y2:\n',Y2)
#####################################################
# Représenter les points de coordonnées X2,Y2 sur le  graphique. 
fig,ax = plt.subplots()
ax.plot(X,Y,marker = 'o',markeredgecolor='blue')
ax.scatter(X2,Y2,marker = '*',color='red')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_xticks(np.arange(0, 10, 1))
#####################################################
# Fusionner d'abord les tableaux X et X2 puis Y et Y2
# Fusion des X
fusionX =np.empty(X.size + X2.size, dtype=X2.dtype)# creer un tableau vide de la taille de X + taille de X2
                                                          # du même type que X2 (pour avoir des décimales)
fusionX[0::2] = X # remplir ce tableau en partant de l'indice 0 et en incrementant l'indice de (indices pairs)
fusionX[1::2] = X2 # remplir ce tableau en partant de l'indice 1 et en incrementant l'indice de (indices impairs)

# Fusion des Y
fusionY =np.empty(Y.size + Y2.size, dtype=X2.dtype)
                                                        
fusionY[0::2] = Y 
fusionY[1::2] = Y2 
#####################################################
# Vérifier en faisant le graphique des tableaux fusionnés
fig,ax = plt.subplots()
ax.plot(fusionX,fusionY,marker = 'o',markeredgecolor='blue')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Fusion')
#####################################################
# Tracer la dérivée des Y fusionnés par rapport aux X fusionnés
# Calcul de la dérivée de y par rapport à x
derivee = np.gradient(fusionY, fusionX)
fig,ax = plt.subplots()
ax.plot(fusionX,fusionY,marker = 'o',markeredgecolor='blue',label = 'Fusion')
ax.plot(fusionX,derivee,label = 'Derivee')
ax.axhline(y= 0, color='grey', linestyle='--')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Fusion et derivee')
ax.legend()
#####################################################
# Trouvez l'ordonnée du point 0.25 par interpolation
ordonnee_point_025 = np.interp(0.25, fusionX, fusionY)
print("L'ordonnée du point 0.25 est :", ordonnee_point_025)

```

- Convertir l'image en niveaux de gris. Une methode est de faire la moyenne des 3 canaux

- Faire un seuillage de l'image. Un seuillage consiste à fixer un seuil s , à partir duquel tout pixel ayant une intensité supérieure ou égale au seuil s sest affecté de la valeur 255 et les autres ayant la valeur 0

```python
# Il n'y a plus qu'un canal
image_seuil = np.copy(image_gray)

seuil = 200
condition_fond = (image_gray < seuil)
condition_objet = (image_gray >= seuil)

indices_fond= np.where(condition_fond)
indices_objet= np.where(condition_objet )

image_seuil[indices_fond] = 0
image_seuil[indices_objet] = 255
```

```python
fig,ax = plt.subplots(1,2)
ax[0].imshow(image_gray,cmap='gray')
ax[0].set_title('Avant')
ax[1].imshow(image_seuil,cmap='gray')#,cmap='gray',vmin = 0, vmax = 255)
ax[1].set_title('Après')
```
