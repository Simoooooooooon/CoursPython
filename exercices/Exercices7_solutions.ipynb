- Afficher l'histogramme des repartitions par age

```python
import matplotlib.pyplot as plt

# Créer la figure et les axes
fig, ax = plt.subplots()

# Tracer l'histogramme
ax.hist(df['age'], bins=30)

# Ajouter des labels aux axes
ax.set_xlabel('Age')
ax.set_ylabel('Frequency')

# Afficher la figure
plt.show()
```

- Afficher l'ensemble des boxplot

```python

# Créer la figure et les sous-graphiques
fig, axes = plt.subplots(5, 3, figsize=(12, 12))

# Boucle pour tracer les boîtes à moustaches de chaque colonne
for i, column in enumerate(df.columns):
    # Calcul des indices de la position du sous-graphique
    row = i // 3
    col = i % 3
    
    # Tracer la boîte à moustaches de la colonne sur le sous-graphique correspondant
    axes[row, col].boxplot(df[column])
    axes[row, col].set_xlabel(column)
    axes[row, col].set_ylabel('Value')
    axes[row, col].set_title('Boxplot of ' + column)

# Ajuster l'espacement entre les sous-graphiques
plt.tight_layout()

# Afficher la figure
plt.show()
```

- Représenter les correlation entre les paramètres d'abord avec une matrice puis avec des graphiques

```python
import seaborn as sns

# Calculer la matrice de corrélation
corr_matrix = df.corr()

# Créer une figure et un axe pour le tracé
fig, ax = plt.subplots(figsize=(10, 8))

# Tracer la matrice de corrélation avec heatmap
sns.heatmap(corr_matrix, cmap='coolwarm', annot=True, ax=ax)

# Configurer les étiquettes des axes
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, horizontalalignment='right')

# Afficher la figure
plt.show()
```

- Slope et oldpeak sembles fortement corrélés. Peut être parce que sont tous les deux des parametres mésurés lors d'un exercice.
- fbs est trés peu corrélé à l'apparition de maladie. C'est normal c'est une condition de msure (examen à jeun ou non)

```python
# Tracer la matrice de scatter plots
pd.plotting.scatter_matrix(df, figsize=(12, 12))

# Ajuster les espacements entre les sous-graphiques
plt.tight_layout()

# Afficher le graphique
plt.show()
```

```python
import seaborn as sns
from matplotlib import pyplot as plt

# Calculer la matrice de corrélation il faut préciser que l'on ne pourra utiliser que les valeurs nomeriques
corr_matrix = df.corr(numeric_only=True)

# Créer une figure et un axe pour le tracé
fig, ax = plt.subplots(figsize=(10, 8))

# Tracer la matrice de corrélation avec heatmap
sns.heatmap(corr_matrix, cmap='coolwarm', annot=True, ax=ax)

# Configurer les étiquettes des axes
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, horizontalalignment='right')

# Afficher la figure
plt.show()
```

```python
df.columns.values 
```

```python
# Grouper les données par Sous-groupe d'aliment et calculer la moyenne pour chaque groupe
grouped_data = df.groupby("Sous-groupe d'aliment")["Changement climatique"].mean().reset_index()

fig,ax = plt.subplots(figsize = (8,10)) 
ax.plot(grouped_data["Sous-groupe d'aliment"],grouped_data["Changement climatique"])

ax.set_xticks(range(len(grouped_data["Sous-groupe d'aliment"])))
ax.set_xticklabels(grouped_data["Sous-groupe d'aliment"], rotation=90)  ;
```

```python

# Grouper les données par Sous-groupe d'aliment et calculer la moyenne pour chaque groupe
climat_data = df.groupby("Groupe d'aliment")["Changement climatique"].mean().reset_index()
rayonnements_data = df.groupby("Groupe d'aliment")["Rayonnements ionisants"].mean().reset_index()

fig,ax = plt.subplots() 
ax.plot(climat_data["Groupe d'aliment"],climat_data["Changement climatique"], label = "Changement climatique")
ax.plot(rayonnements_data["Groupe d'aliment"],rayonnements_data["Rayonnements ionisants"], label = "Rayonnements ionisants")
ax.legend()

ax.set_xticks(range(len(climat_data["Groupe d'aliment"])))
ax.set_xticklabels(climat_data["Groupe d'aliment"], rotation=90)  
;
```

```python
columns_to_plot = ['Changement climatique', "Appauvrissement de la couche d'ozone", 'Rayonnements ionisants',
                   "Formation photochimique d'ozone", 'Particules fines',
                   'Effets toxicologiques sur la santé humaine\xa0: substances non-cancérogènes',
                   'Effets toxicologiques sur la santé humaine\xa0: substances cancérogènes',
                   'Acidification terrestre et eaux douces', 'Eutrophisation eaux douces',
                   'Eutrophisation marine', 'Eutrophisation terrestre',
                   "Écotoxicité pour écosystèmes aquatiques d'eau douce", 'Utilisation du sol',
                   'Épuisement des ressources eau', 'Épuisement des ressources énergétiques',
                   'Épuisement des ressources minéraux']




# Calculer le minimum et le maximum de chaque colonne
for column in df[columns_to_plot].values:
    min_value = column.min()
    max_value = column.max()

# Appliquer la transformation aux données
for column in df[columns_to_plot].values:
    column = (column - min_value) / (max_value - min_value)

# Afficher les données normalisées
df[columns_to_plot]
```

```python
fig, ax = plt.subplots()

# Boucle sur les colonnes pour tracer les données
for column in columns_to_plot:
    data = df.groupby("Groupe d'aliment")[column].mean().reset_index()
    ax.plot(data["Groupe d'aliment"], data[column], label=column)

ax.legend()
ax.set_xticks(range(len(data["Groupe d'aliment"])))
ax.set_xticklabels(data["Groupe d'aliment"], rotation=90)
ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
# Obtenir toutes les lignes dont la colonne ingredient contient pomme
df_pomme = df[df["Ingredients"].str.contains("Pomme")]
# Afficher les résultats
df_pomme
```

```python


# Obtenir la liste des sous-groupes d'aliments
sous_groupes = df["Sous-groupe d'aliment"].unique()

# Calculer l'impact du changement climatique pour chaque sous-groupe
impacts_cc = []
for sous_groupe in sous_groupes:
    impacts_cc.append(df[df["Sous-groupe d'aliment"] == sous_groupe]["Changement climatique"].mean())

# Créer le graphique
plt.bar(sous_groupes, impacts_cc)
plt.xlabel("Sous-groupe d'aliment")
plt.ylabel("Changement climatique")
plt.xticks(rotation=90)
plt.show()
```
